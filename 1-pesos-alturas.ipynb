# Matplotlib

```python
from matplotlib import pyplot as plt
```

```python
data_hombres = data[data['Genero']=='Hombre'][['Peso','Altura']].values
print(data_hombres)
```

```python
data_mujeres = data[data['Genero']=='Mujer'][['Peso','Altura']].values
print(data_mujeres)
```

```python
f = plt.figure(figsize=(20,10))
plt.scatter(data_hombres[:,0], data_hombres[:,1], color='b', s=2, label='Hombres')
plt.scatter(data_mujeres[:,0], data_mujeres[:,1], color='r', s=2, label='Mujeres')
plt.xlabel('Pesos [cms]')
plt.ylabel('Alturas [cms]')
plt.legend()
plt.show()
```
